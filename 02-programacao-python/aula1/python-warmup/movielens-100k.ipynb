### 1. Most critical users

```python
data.groupby(by='user id').mean().drop(['item id', 'timestamp'], axis=1).reset_index(level=0)
```

```python
# Def a function that give the top n critical users
def top_critical_users(n):
    pd.set_option('display.max_rows', n)
    tcu=data.groupby(by='user id').mean().drop(['item id', 'timestamp'], axis=1).sort_values(by='rating').reset_index(level=0).head(n)
    return(tcu)
```

```python
top_critical_users(5)
```

### 2. Worst rated movies

```python
# Def a function that gives the top n worst rated movies
def top_worse_movies(n):
    pd.set_option('display.max_rows', n)
    twm=data.groupby(by='item id').mean().drop(['user id', 'timestamp'], axis=1).sort_values(by='rating').reset_index(level=0).head(n)
    return(twm)
```

```python
top_worse_movies(71)
```

### 3. Best rated movies

```python
# Def a function that give the top n best rated movies
def top_best_movies(n):
    pd.set_option('display.max_rows', n)
    tbm=data.groupby(by='item id').mean().drop(['user id', 'timestamp'], axis=1).sort_values(by='rating', ascending=False).reset_index(level=0).head(n)
    return(tbm)
```

```python
top_best_movies(11)
```

### 4. Rating by genre

#### Preprocessing data frames

```python
# Drop useless column from u.item
dcols = ['movie title',
        'release date',
        'video release date',
        'IMDb URL',
        ]
d_item = item.drop(dcols, axis=1)
d_item
```

```python
# Get the mean rating by movies('item id')
d_data = data.groupby(by='item id').mean().drop(['user id', 'timestamp'], axis=1).sort_values(by='item id').reset_index(level=0)
d_data
```

```python
# Concatenate d_data and d_item
data_item = pd.concat([d_data, d_item], axis=1).drop(['item id'], axis=1)
data_item
```

#### Creating a dictionary

```python
data_item.columns
```

```python
genre = ['unknown',
         'Action',
         'Adventure', 
         'Animation',
         "Children's",
         'Comedy',
         'Crime',
         'Documentary', 
         'Drama', 
         'Fantasy', 
         'Film-Noir',
         'Horror', 
         'Musical', 
         'Mystery', 
         'Romance', 
         'Sci-Fi', 
         'Thriller', 
         'War',
         'Western']
```

```python
# Add the rating by genre on a dictionary
gen_mean={}
for gen in genre:
    gen_mean[gen]=round(data_item.loc[lambda data_item: data_item[gen] == 1]['rating'].mean(),1)
```

```python
gen_mean
```

### 5. Mean rating by year

Preprocessing data frames

```python
item['year']=pd.DatetimeIndex(item['release date']).year
item['year']
```

```python
rating_year = pd.concat([d_data, item['year']], axis=1).drop(['item id'], axis=1)
rating_year
```

Defining a function

```python
# Define a function that gives the top n best rated years
def top_critical_years(n):
    pd.set_option('display.max_rows', n)
    tcy=rating_year.groupby('year').mean().sort_values('rating', ascending=False).reset_index(level=0).head(n)
    return(tcy)
```

```python
top_critical_years(5)
```

### 6. Top critical occupation

Preprocessing data frames

```python
# Concatenate d_data with only de relevant columns from u.user
data_user = pd.concat([d_data, user[['user id', 'occupation']]], axis=1).drop(['user id', 'item id'], axis=1)
data_user
```

Defining a function

```python
# Define a function that gives the top n best rated years
def top_critical_occupations(n):
    pd.set_option('display.max_rows', n)
    tco=data_user.groupby('occupation').mean().sort_values('rating').reset_index(level=0).head(n)
    return(tco)
```

```python
top_critical_occupations(5)
```
